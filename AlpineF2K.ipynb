2. Donald Trump is 6ft 3in tall and weights 244 pounds. Barack Obama is 6ft 1in and  179.3 pounds. What are their heights in metric?

```python
#1 foot = 30.48cm
#1 inch = 2.54cm

def height_imperial2metric(foot,inch):
    
    # complete here
    height=0
    height=foot*30.48+inch+2.54
    return height

print(height_imperial2metric(6,3))
print(height_imperial2metric(6,1))
```

3. Based on BMI formula and categories, write a function that takes a person's height (cm), and weight (kg), computes BMI, then tells which BMI category he/she belongs to?

```python
def compute_bmi(name,height_cm,weight_kg):
    height_m = 0.01 * height_cm
    bmi=weight_kg/(height_m*height_m)
    if bmi <= 18.5:
        print(f"{name} is underweight: bmi {bmi}")
    elif bmi > 18.5 and bmi < 24.9:
        print(f"{name} is healthy: bmi {bmi}")
    elif bmi > 25 and bmi < 29.9:
        print(f"{name} is overweight: bmi {bmi}")
    else:
        print(f"{name} is obese: bmi {bmi}")
        
compute_bmi("Donald Trump",188.42,110.91)
compute_bmi("Barack Obama",186.42,81.5)
```
